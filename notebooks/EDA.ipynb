```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
sns.set(style="whitegrid")

file_path = '../data/raw/Students Social Media Addiction.csv'
df = pd.read_csv(file_path)

display(df.head())
```

```python
print("\n--- Informasi Dataset ---")
df.info()

print("\n--- Statistik Deskriptif ---")
display(df.describe())

print("\n--- Pengecekan Data Kosong ---")
print(df.isnull().sum())
```

```python
fig_path = '../outputs/figures/'

# 1. Distribusi Platform yang Paling Sering Digunakan
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Most_Used_Platform', palette='viridis', order=df['Most_Used_Platform'].value_counts().index)
plt.title('Distribusi Platform Media Sosial Paling Dominan')
plt.xlabel('Platform')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)
plt.savefig(os.path.join(fig_path, 'distribution_platform.png'))
plt.show()
```

```python
# 2. Distribusi Durasi Penggunaan Harian
plt.figure(figsize=(10, 6))
sns.histplot(df['Avg_Daily_Usage_Hours'], bins=15, kde=True, color='royalblue')
plt.title('Sebaran Durasi Penggunaan Media Sosial Harian')
plt.xlabel('Jam per Hari')
plt.ylabel('Frekuensi')
plt.savefig(os.path.join(fig_path, 'distribution_usage_hours.png'))
plt.show()
```

```python
# 5. Heatmap Korelasi
plt.figure(figsize=(12, 10))
# Mengambil hanya kolom numerik untuk korelasi
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel Perilaku Digital')
plt.savefig(os.path.join(fig_path, 'correlation_heatmap.png'))
plt.show()
```
