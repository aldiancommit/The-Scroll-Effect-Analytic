```python
import pandas as pd
import joblib
import os
from sklearn.preprocessing import LabelEncoder

# 1. Load data dari folder raw
file_path = '../data/raw/Students Social Media Addiction.csv'
df = pd.read_csv(file_path)
```

```python
# MENGHAPUS VARIABEL LEAKAGE & IDENTITAS
# Hapus Mental_Health_Score karena ini adalah 'gejala', bukan 'penyebab'
cols_to_drop = ['Student_ID', 'Mental_Health_Score']
df_cleaned = df.drop(columns=cols_to_drop)

print(f"Variabel yang digunakan untuk prediksi: {df_cleaned.columns.tolist()}")
```

```python
# Daftar kolom yang berisi teks (Kategori)
categorical_cols = [
    'Gender', 'Most_Used_Platform', 'Academic_Level', 
    'Relationship_Status', 'Country', 'Affects_Academic_Performance'
]

# Folder untuk menyimpan encoder
encoder_folder = '../models/encoders/'
if not os.path.exists(encoder_folder):
    os.makedirs(encoder_folder)
```

```python
# Proses Encoding
for col in categorical_cols:
    le = LabelEncoder()
    # Pastikan data berupa string agar tidak error
    df_cleaned[col] = le.fit_transform(df_cleaned[col].astype(str))
    
    # SIMPAN ENCODER: Ini 'kunci' untuk aplikasi web nanti
    joblib.dump(le, f"{encoder_folder}{col}_encoder.pkl")
    print(f"Berhasil mengonversi dan menyimpan encoder untuk: {col}")

print("\nSemua data teks telah diubah menjadi angka yang dipahami mesin.")
```

```python
# Simpan data yang sudah bersih
processed_path = '../data/processed/cleaned_data.csv'
df_cleaned.to_csv(processed_path, index=False)

print(f"Data bersih berhasil disimpan di: {processed_path}")
print(f"Ukuran Data: {df_cleaned.shape}")
```
